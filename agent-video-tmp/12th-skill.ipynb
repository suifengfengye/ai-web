# 第12期：Skill 和 Rule 一页讲清楚

这份说明用最简单的方式解释：什么是 Skill、什么是 Rule，以及它们在本项目里如何配合使用。

## 1. 先用一句话理解

- **Skill**：告诉 AI **怎么做一件事**（执行流程/SOP）。
- **Rule**：告诉 AI **必须遵守什么**（长期约束/红线）。

记忆法：
- Skill 像“操作手册”。
- Rule 像“公司制度”。

## 2. Skill 和 Rule 的区别

| 维度 | Skill | Rule |
|---|---|---|
| 作用 | 提供步骤化做法 | 提供必须遵守的约束 |
| 关注点 | “怎么做” | “必须做/不能做” |
| 触发方式 | 根据任务语义自动触发（或被明确提及） | 项目范围内持续生效 |
| 适合内容 | 验证流程、发布流程、排障流程 | 环境约束、验收标准、输出规范 |
| 违反后果 | 可能漏步骤 | 违反项目规范，结果不合格 |

## 3. 结合本项目举例

### 3.1 你的 Skill（做法）

文件：`.trae/skills/agent-video-verify/SKILL.md`

它描述了：
- 当代码改动后，要去对应 `数字th-` 目录验证。
- 在 conda 环境 `agent-video` 中运行 `python agent.py`。
- 如果 conda 激活有问题，用 `conda run -n agent-video python agent.py`。

这就是“**怎么验证**”的标准流程。

### 3.2 你的 Rule（约束）

文件：`.trae/rules/project_rules.md`

它约束了：
- 只要改了 `*th-*` 目录代码，回复前必须验证。
- 多个目录改动要逐个验证。
- 任一目录失败不能宣告完成。
- 最终反馈必须写清：目录、命令、成功/失败。

这就是“**必须做到什么程度才算完成**”。

## 4. 两者如何配合

- Rule 决定：**这件事必须做**（不能跳过验证）。
- Skill 决定：**这件事怎么做**（按哪套命令和步骤验证）。

一句话：
> Rule 管“底线”，Skill 管“方法”。

## 5. 对外讲解模板（可直接复用）

你可以这样对同事说：

> 我们项目里，Rule 负责规定验收标准：改了 `th-` 目录必须跑 `agent.py` 验证，失败不能交付；  
> Skill 负责给出验证步骤：进入对应目录，在 `agent-video` 环境执行 `python agent.py`，并汇总结果。  
> 两者结合后，AI 既知道“必须做”，也知道“怎么做”。

## 6. 你这个项目的最佳实践

- 保留当前 `agent-video-verify` Skill：沉淀验证流程。
- 保留当前 `project_rules.md`：强制执行验证标准。
- 以后如果新增“发布流程”“回归测试流程”，优先新增 Skill；
- 如果新增“必须满足的验收条件”，写进 Rule。
