4. HTML URL: https://api.wikimedia.org/core/v1/wikipedia/zh/page/{XXX}/html

通过 wikipedia 返回的内容是静态的HTML的内容。
1. wikipedia是以结构化(wikitext)的语言存储。
2. 当请求某一个词条的时候，wikipedia会将wikitext渲染成HTML页面返回。
3. 缓存：为了减轻服务器压力，生成的HTML文件会被缓存起来，下一次访问时直接返回HTML即可，不需要再次渲染。
4. CDN: 同时也会将HTML文件分发到全球的CDN网络，就近返回给用户。

```python
import requests

language_code = 'zh'
search_query = '成都'
number_of_results = 3
headers = {
  # 'Authorization': 'Bearer YOUR_ACCESS_TOKEN',
  'User-Agent': 'DXH_APP'
}

base_url = 'https://api.wikimedia.org/core/v1/wikipedia/'
endpoint = '/search/page'
url = base_url + language_code + endpoint
parameters = {'q': search_query, 'limit': number_of_results}
response = requests.get(url, headers=headers, params=parameters)
```
