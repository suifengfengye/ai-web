# 1. 流式输出

文档链接：https://docs.langchain.com/oss/python/langchain/streaming/overview#accessing-completed-messages

create_agent()方法设置流式传输的参数:
- stream_mode: updates|messages|custom,3个取值
    - updates: 在agent的每个步骤执行后，流式地传输状态更新。
    - messages：LLM的流式输出
    - custom：自定义传输流式数据，一般是在@tool节点内部使用了流式数据传输的方式时会使用到。
