2. 静态方法

- from_list(): 将一个列表转换成数据集。
- from_dict(): 将一个字典转换成数据集。(实际上是一个字典列表，和from_list一致)
- from_jsonl(path: str | Path): 将jsonl文件转换成数据集。
    - JSONL 文件 （JSON Lines），则是一种每行包含一个独立的JSON 对象的文本文件格式。 


````python
from ragas import EvaluationDataset
dataset = EvaluationDataset.from_jsonl("questions_answers_01.txt")
dataset.to_pandas()
````

# 2. evaluate()

使用不同的指标在一个数据集上进行评估。

```python
from ragas import evaluate
```

参数:

- dataset: 数据集
- metrics: 指标列表
- llm: 大语言模型实例，经过 LangchainLLMWrapper 包装
- run_config: RunConfig对象，评估过程的配置信息，
    - 如：timeout(180s,超时时间) / max_retries(10, 最大重试次数) / max_workers（16, 工作线程池）

````python
from ragas import evaluate, RunConfig
from ragas.metrics import FactualCorrectness
from langchain_ollama import ChatOllama
from ragas.llms import LangchainLLMWrapper

llm = ChatOllama(model="qwen2.5:latest")

metric = FactualCorrectness(
    atomicity="low",
    coverage="high",
    llm=LangchainLLMWrapper(llm),
)

result = evaluate(
    dataset=dataset,
    metrics=[metric],
    llm=llm,
    run_config=RunConfig(
        timeout=600,
        max_retries=3,
        max_workers=4
    ),
)
result
````
